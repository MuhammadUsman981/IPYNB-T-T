# 16- Spliting a string into multiple columns

```python
import pandas as pd

df = pd.DataFrame({'name':['Muhammad Usman','Muhammad Adil','Sohail Usman','Nafeel Khan'],
                  'location':['Peshawar, Pakistan','Akbarpura, pakistan','Bannu, Pakistan','Punjab, Pakistan'] })
df 
```

```python
# Splitting the columns 
df.name.str.split(' ', expand = True)
```

```python
# Adding those splits to the New column 
df[['first_name','last_name']]=df.name.str.split(' ', expand = True)
```

```python
df
```

```python
df.location.str.split(', ',expand=True)
```

```python
df[["City","Country"]] = df.location.str.split(', ',expand=True)
df
```

```python
# Refine data manipulation
df = df[['first_name','last_name','City','Country']]
df
```
